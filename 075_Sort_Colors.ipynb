# 75. Sort Colors
Given an array with n objects colored red, white or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white and blue.

Here, we will use the integers 0, 1, and 2 to represent the color red, white, and blue respectively.

Note: You are not suppose to use the library's sort function for this problem.
<div class="alert alert-block alert-info">
Example:

Input: [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]
Follow up:

A rather straight forward solution is a two-pass algorithm using counting sort.
First, iterate the array counting number of 0's, 1's, and 2's, then overwrite array with total number of 0's, then 1's and followed by 2's.
Could you come up with a one-pass algorithm using only constant space?
</div>


## Topic analysis
Combine the application of the three-way fast-distribution partition idea ( or This is a dutch partitioning problem).

Set two indexes, one to slide from left to right "zero_end" and one to slide from right to left "two_end".

* Traversing nums, when nums[i] the value is 1 i++,
* When nums[i] the value is 2, two the value is first decremented by 1, and then exchanged nums[i] with the value nums[two]observed at this time nums[i];
* When nums[i]0 value, zero++then the exchange nums[i]with nums[zero], ; i++ when i = two the loop ends.



```python
def sortColors(nums):
    """
    Do not return anything, modify nums in-place instead.
    """
    zero_end, two_end = 0,len(nums)-1
    cur = 0
    while cur <= two_end:
        if nums[cur] == 0:
            nums[cur],nums[zero_end] = nums[zero_end],nums[cur]
            zero_end +=1
            cur +=1
        elif nums[cur] == 2:
            nums[cur],nums[two_end] = nums[two_end],nums[cur]
            two_end -=1
        else:
            cur +=1

A = [2,0,2,1,1,0]
sortColors(A)
A
```
