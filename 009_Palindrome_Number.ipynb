Determine whether an integer is a palindrome. An integer is a palindrome when it reads the same backward as forward.Coud you solve it without converting the integer to a string?
<div class="alert alert-block alert-info">
Example 1:

Input: 121
Output: true
Example 2:

Input: -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

Solution 1: Mathematical Solution
The rounding and remainder operations are used to obtain the corresponding numbers in the integers for comparison.

For example: 1221 this number.

By calculating 1221 / 1000, the first place is 1
By calculating 1221% 10, you can get the last bit 1
Compare
Then take 22 out and continue to compare

```python
def isPalindrome(self, x: int) -> bool:
    #reverse the number first and see if it is equal to the original number.
    if x < 0:
        return False
    p,res = x, 0
    while p:
        res = res * 10 + p % 10
        p = p//10
    return res == x

```
