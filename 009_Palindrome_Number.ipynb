## Solution 1: Mathematical Solution
The rounding and remainder operations are used to obtain the corresponding numbers in the integers for comparison.

For example: 1221 this number.

By calculating 1221 / 1000, the first place is 1
By calculating 1221% 10, you can get the last bit 1

Then take 22 out and continue to compare

```python
def isPalindrome(x: int) -> bool:
    #reverse the number first and see if it is equal to the original number.
    if x < 0:
        return False
    cur,res = x, 0
    while cur:
        res = res * 10 + cur % 10
        cur = cur//10
    return res == x

def isPalindrome_2(x: int) -> bool:
    if x < 0:
        return False
    #find the largest power of 10 which is smaller than x
    largestPowerOfTen = 1
    while x//largestPowerOfTen >=10:
        largestPowerOfTen *=10
    #check if left most digit is equal to right most digit and so on.
    while largestPowerOfTen >= 10:
        left,right = x//largestPowerOfTen, x%largestPowerOfTen
        if left != right:
            return False
        x,largestPowerOfTen = (x%largestPowerOfTen)//10, largestPowerOfTen//10
    
    return True
    
```
