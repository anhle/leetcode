## 37. Sudoku Solver
Write a program to solve a Sudoku puzzle by filling the empty cells.

A sudoku solution must satisfy all of the following rules:

* Each of the digits 1-9 must occur exactly once in each row.
* Each of the digits 1-9 must occur exactly once in each column.
* Each of the the digits 1-9 must occur exactly once in each of the 9 3x3 sub-boxes of the grid.
* Empty cells are indicated by the character '.'.




Note:

* The given board contain only digits 1-9 and the character '.'.
* You may assume that the given Sudoku puzzle will have a single unique solution.
* The given board size is always 9x9.

```python
def solveSudoku(board):
    #Use three sets to track the available choices for each row/column/square.
    row_remain = [set(range(1,10)) for _ in range(9)]
    col_remain = [set(range(1,10)) for _ in range(9)]
    square_remain = [set(range(1,10)) for _ in range(9)]
    # find available remain choices
    count = 0
    for i in range(9):
        for j in range(9):
            if board[i][j] != '.':
                val = int(board[i][j])
                row_remain[i].remove(val)
                col_remain[j].remove(val)
                square_remain[int(3*(i//3) + j//3)].remove(val)
            else:
                count +=1

    #back track search
    def dfs(i,j,count):
        if count == 0:
            return True
        # find the first blank
        while board[i][j] != '.':
            if j == 8:
                i +=1
                j = 0
            else:
                j +=1
        available = row_remain[i] & col_remain[j] & square_remain[int(3*(i//3)+j//3)]
        for choice in available:
            board[i][j] = str(choice)
            row_remain[i].remove(choice)
            col_remain[j].remove(choice)
            square_remain[int(3*(i//3)+j//3)].remove(choice)
            if dfs(i,j,count-1):
                return True
            row_remain[i].add(choice)
            col_remain[j].add(choice)
            square_remain[int(3*(i//3)+j//3)].add(choice)
            board[i][j] = '.'
        return False

    dfs(0,0,count)
    print(board)
    
solveSudoku([["5","3",".",".","7",".",".",".","."],
             ["6",".",".","1","9","5",".",".","."],
             [".","9","8",".",".",".",".","6","."],
             ["8",".",".",".","6",".",".",".","3"],
             ["4",".",".","8",".","3",".",".","1"],
             ["7",".",".",".","2",".",".",".","6"],
             [".","6",".",".",".",".","2","8","."],
             [".",".",".","4","1","9",".",".","5"],
             [".",".",".",".","8",".",".","7","9"]])
```
